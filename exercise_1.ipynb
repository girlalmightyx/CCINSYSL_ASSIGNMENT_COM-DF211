```python
# Answer here
def lowercase(transform):
  print(transform.lower())

lowercase('CAT')
```

2. Find the function **f(x)**  

| Input  | CAT | DOG | APPLE | BANANA | ORANGE |  
| --- | --- | --- | --- | --- | --- |
| Output | CT | DG | PPL | BNN | RNG


```python
def rem_vowel(string):
    vowels = ['a','e','i','o','u']
    result = [letter for letter in string if letter.lower() not in vowels]
    result = ''.join(result)
    print(result)
rem_vowel('CAT')
```

5. Find the function **f(x)**  

| Input  | 64 | 27 | 1000 | 729 | 512 |  
| --- | --- | --- | --- | --- | --- |
| Output | 4 | 3 | 10 | 9 | 8

```python
def cube(number):
  cube_root = number ** (1/3)
  rounded_cube_root = round(cube_root)
  print(str(rounded_cube_root))
```

```python
number=512
cube(number)
```
